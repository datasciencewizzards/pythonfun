# Pivot and Pivot table 

```python
# Setup enviroment
# import libraries
import pandas as pd
import numpy as np

df = pd.read_csv("weather.csv")
df
```

```python
pivot_df = df.pivot(index='humidity',columns='city')
pivot_df
```

```python
# Pivot table let's you run aggregate fuctions ; it is better called aggregated table
```

```python
df = pd.read_csv('weather2.csv')
df.pivot_table(index='city',columns='date',margins=True)
```

```python
# pivot table grouper
```

```python
df = pd.read_csv('weather2.csv')
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```

```python
#  To fix the above error; convert the string date to datetime
df.date = pd.to_datetime(df.date)
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
