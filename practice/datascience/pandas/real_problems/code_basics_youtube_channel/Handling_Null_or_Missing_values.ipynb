```python
# We can see the type of the day is str
# We can load the str as date
# Read CSV
df = pd.read_csv('weather_data_missing_vals.csv',parse_dates=['day'])
df
```

```python
# Check the type now, it mus be Timestamp
type(df.day[0])
```

```python
# Replace all NA values with 0
new_df = df.fillna(0)
new_df
```

```python
# Shoot : Event contains 0 now, mis-matched datatype
# Pass a dictionary to fillna
new_df = df.fillna({
    'temperature': 0,
    'windspeed' : 0,
    'event':'No Event'
})

new_df
```

```python
# Fill with carry forward value from the previous entry
new_df = df.fillna(method='ffill') # ffill : forwardfill
new_df
```

```python
# Fill with carry forward value from the previous entry
new_df = df.fillna(method='bfill') # ffill : backwardfill
new_df
```

```python
# Fill columns wise with carry forward value from the previous entry
new_df = df.fillna(method='bfill',axis="columns") # ffill : backwardfill
new_df
```

```python
# Fill with carry forward value from the previous entry and limiting the forward fill
new_df = df.fillna(method='ffill',limit=1) # ffill : forwardfill
new_df
```

```python
# Fill with interplolated value
new_df = df.interpolate() # Fills with with the mean , applicable for only the numeric values
new_df
```

```python
# Earlier interploation was on the linear but if we check the dates the dates were not linear
# See if something can be done
# df.set_index('day',inplace=True)
new_df = df.interpolate('time') # Fills with with the mean , applicable for only the numeric values
#type(new_df.day[0])
new_df
```

```python
# Dropping NA Values
```

```python
new_df= df.dropna()
new_df
```

```python
# Drop only if all are NA
new_df= df.dropna(how="all")
new_df
```
